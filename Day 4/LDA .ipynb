### Company Bankruptcy

```python
import pandas as pd
df = pd.read_csv('../Cases/Company Bankruptcy/data.csv')
df.head()
```

```python
# Separating Features
X, y = df.drop('Bankrupt?', axis=1), df['Bankrupt?']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24, stratify=y)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
y_pred = lda.predict(X_test)
print('accuracy_score', accuracy_score(y_test, y_pred))
```

### Glass Identification

```python
glass = pd.read_csv('../Cases/Glass Identification/Glass.csv')
X, y = glass.drop('Type', axis=1), glass['Type']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24, stratify=y)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
y_pred = lda.predict(X_test)
print('accuracy_score', accuracy_score(y_test, y_pred))
```

### LDA Transformation

```python
len(y.unique())
```

```python
### Transformation using LDA

X_trn_lda = lda.transform(X_train)
X_tst_lda = lda.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
n_list = np.arange(1,11)
scores = []
for n in n_list:
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_trn_lda, y_train)
    y_pred = knn.predict(X_tst_lda)
    scores.append(accuracy_score(y_test, y_pred))

print('good neighbors : ', n_list[np.argmax(scores)])
print('good accuracy : ', np.max(scores))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_trn_lda, y_train)
y_pred = nb.predict(X_tst_lda)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score

# Define the pipeline with LDA and KNN
pipe = Pipeline([
    ('lda', LinearDiscriminantAnalysis()),  # LDA for dimensionality reduction
    ('knn', KNeighborsClassifier(n_neighbors=8))  # KNN for classification
])
pipe.fit(X_trn_lda, y_train)
y_pred = pipe.predict(X_tst_lda) 

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
