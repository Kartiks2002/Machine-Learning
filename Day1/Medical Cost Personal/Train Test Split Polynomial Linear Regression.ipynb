```python
import pandas as pd
df = pd.read_csv('insurance.csv')
df
```

```python
X = df.drop('charges', axis=1)
y = df['charges']
```

### One hot encoding

```python
X_ohe = pd.get_dummies(X, drop_first=True)
X_ohe
```

### With degree 1

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_ohe, y, test_size=0.3, random_state=24)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f'R-Squared : {r2}')
```

### With Degree 2

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X_ohe)
len(poly.get_feature_names_out())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.3, random_state=24)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f'R-Squared : {r2}')
```

### The Value of R2 increased from 0.76 to 0.84 by using degree 2
