```python
import pandas as pd
df = pd.read_csv('JobSalary2.csv')
df
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='constant', fill_value=50) # returns ndarray
imp.fit_transform(df) # ndarray
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='constant', fill_value=50).set_output(transform='pandas') # returns a dataframe
imp.fit_transform(df) # df
```

```python
### Mean Strategy
imp = SimpleImputer(strategy='mean').set_output(transform='pandas')
imp.fit_transform(df)
```
