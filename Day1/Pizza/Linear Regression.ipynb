```python
import pandas as pd
df = pd.read_csv('pizza.csv')
df
```

```python
import matplotlib.pyplot
matplotlib.pyplot.figure(figsize=(10,5))
matplotlib.pyplot.scatter(df.Promote, df.Sales)
```

```python
X = df[['Promote']]
y = df['Sales']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
print('b0 = ', model.intercept_)
print('b1 = ', model.coef_)
```

```python
promo = [100,200,300,400,500]
sales = []
```

```python
for i in range(5):
    print(f'Promo = {promo[i]} : ', model.predict([[promo[i]]]))
```

```python
import numpy as np
tst_promote = np.array([100,200,250,300])
model.intercept_ + tst_promote*model.coef_
```

```python
tst = pd.DataFrame({'Promote' : tst_promote})
tst
```

```python
result = model.predict(tst)
result
```
