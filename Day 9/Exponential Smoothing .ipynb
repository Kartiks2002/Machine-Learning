## Milk Production Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from statsmodels.tsa.api import SimpleExpSmoothing, Holt, ExponentialSmoothing
```

```python
df = pd.read_csv('../Datasets/monthly-milk-production-pounds-p.csv')
y = df['Milk']
y_train, y_test = y.iloc[:-12], y.iloc[-12:]
y_train.shape, y_test.shape
```

### SimpleExpSmoothing

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from sklearn.metrics import mean_squared_error as mse

alpha = 0.1
ses = SimpleExpSmoothing(y_train)
fit1 = ses.fit(smoothing_level=alpha)
fcast1 = fit1.forecast(len(y_test))
y_test.plot(label='Test')
fcast1.plot(label='Forecast')
rmse = np.sqrt(mse(y_test, fcast1))
plt.text(164, 900, "RMSE = " + str(np.round(rmse, 2)))
plt.legend(loc='best')
```

### Holt's Linear Trend

```python
alpha, beta = 0.5, 0.02
holt = Holt(y_train)
fit1 = holt.fit(smoothing_level=alpha, smoothing_trend=beta)
fcast1 = fit1.forecast(len(y_test))
y_test.plot(label='Test')
fcast1.plot(color = 'red', label='Forecast')
rmse = np.sqrt(mse(y_test, fcast1))
plt.text(164, 900, "RMSE = " + str(np.round(rmse, 2)))
plt.legend(loc='best')
```

### ExponentialSmoothing's Linear Trend

```python
alpha, beta = 0.5, 0.02
es = ExponentialSmoothing(y_train)
fit1 = es.fit(smoothing_level=alpha, smoothing_trend=beta)
fcast1 = fit1.forecast(len(y_test))
y_test.plot(label='Test')
fcast1.plot(color = 'red', label='Forecast')
rmse = np.sqrt(mse(y_test, fcast1))
plt.text(164, 900, "RMSE = " + str(np.round(rmse, 2)))
plt.legend(loc='best')
```

### Holt's Exponential

```python
alpha, beta = 0.5, 0.02
holt = Holt(y_train, exponential=True)
fit1 = holt.fit()
fcast1 = fit1.forecast(len(y_test))
y_test.plot(label='Test')
fcast1.plot(color = 'red', label='Forecast')
rmse = np.sqrt(mse(y_test, fcast1))
plt.text(164, 900, "RMSE = " + str(np.round(rmse, 2)))
plt.legend(loc='best')
```

### Additive Damped Trend

```python
alpha, beta, phi = 0.5, 0.02, 0.01
holt = Holt(y_train)
fit1 = holt.fit(smoothing_level=alpha, smoothing_trend=beta, damping_trend=phi)
fcast1 = fit1.forecast(len(y_test))
y_test.plot(label='Test')
fcast1.plot(color = 'red', label='Forecast')
rmse = np.sqrt(mse(y_test, fcast1))
plt.text(164, 900, "RMSE = " + str(np.round(rmse, 2)))
plt.legend(loc='best')
```

### Multiplicative Damped Trend

```python
alpha, beta, phi = 0.5, 0.02, 0.01
holt = Holt(y_train, exponential=True)
fit1 = holt.fit(smoothing_level=alpha, smoothing_trend=beta, damping_trend=phi)
fcast1 = fit1.forecast(len(y_test))
y_test.plot(label='Test')
fcast1.plot(color = 'red', label='Forecast')
rmse = np.sqrt(mse(y_test, fcast1))
plt.text(164, 900, "RMSE = " + str(np.round(rmse, 2)))
plt.legend(loc='best')
```

### Holt-Winters' Additive Method

```python
alpha, beta, gamma = 0.5, 0.02, 0.6
es = ExponentialSmoothing(y_train, seasonal_periods=12, trend='add', seasonal='add')
fit1 = es.fit(smoothing_level=alpha, smoothing_trend=beta, smoothing_seasonal=gamma)
fcast1 = fit1.forecast(len(y_test))
y_test.plot(label='Test')
fcast1.plot(color = 'red', label='Forecast')
rmse = np.sqrt(mse(y_test, fcast1))
plt.text(164, 900, "RMSE = " + str(np.round(rmse, 2)))
plt.legend(loc='best')
```

### Holt-Winters' Multiplicative Method

```python
alpha, beta, gamma = 0.5, 0.02, 0.6
es = ExponentialSmoothing(y_train, seasonal_periods=12, trend='mul', seasonal='add')
fit1 = es.fit(smoothing_level=alpha, smoothing_trend=beta, smoothing_seasonal=gamma)
fcast1 = fit1.forecast(len(y_test))
y_test.plot(label='Test')
fcast1.plot(color = 'red', label='Forecast')
rmse = np.sqrt(mse(y_test, fcast1))
plt.text(164, 900, "RMSE = " + str(np.round(rmse, 2)))
plt.legend(loc='best')
```

## Monthly Traffic Fatalities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from statsmodels.tsa.api import SimpleExpSmoothing, Holt, ExponentialSmoothing
```

```python
df = pd.read_csv('../Datasets/monthly-traffic-fatalities-in-on.csv',delimiter=';')
y = df['fatalities']
y_train, y_test = y.iloc[:-12], y.iloc[-12:]
y_train.shape, y_test.shape
```

### Holt-Winters' Additive Method

```python
alpha, beta, gamma = 0.5, 0.02, 0.6
es = ExponentialSmoothing(y_train, seasonal_periods=12, trend='add', seasonal='add')
fit1 = es.fit(smoothing_level=alpha, smoothing_trend=beta, smoothing_seasonal=gamma)
fcast1 = fit1.forecast(len(y_test))
y_test.plot(label='Test')
fcast1.plot(color = 'red', label='Forecast')
rmse = np.sqrt(mse(y_test, fcast1))
plt.text(168, 200, "RMSE = " + str(np.round(rmse, 2)))
plt.legend(loc='best')
```

### Holt-Winters' Multiplicative Method¶

```python
alpha, beta, gamma = 0.6, 0.2, 0.6
es = ExponentialSmoothing(y_train, seasonal_periods=12, trend='mul', seasonal='mul')
fit1 = es.fit(smoothing_level=alpha, smoothing_trend=beta, smoothing_seasonal=gamma)
fcast1 = fit1.forecast(len(y_test))
y_test.plot(label='Test')
fcast1.plot(color = 'red', label='Forecast')
rmse = np.sqrt(mse(y_test, fcast1))
plt.text(168, 200, "RMSE = " + str(np.round(rmse, 2)))
plt.legend(loc='best')
```
