## Milk Production

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('../Datasets/monthly-milk-production-pounds-p.csv', index_col =0)
```

```python
df.index = pd.to_datetime(df.index).to_period("M")
```

```python
df.plot()
plt.title("Monthly Milk Production")
plt.xlabel("Months")
plt.show()
```

### Quarterly

```python
downsampled = df.resample('Q').sum()
downsampled.index.rename('Quarter',inplace = True)  
downsampled.plot()
plt.title("Monthly Milk Production")
plt.xlabel("Months")
plt.show()
```

### Yearly

```python
downsampled = df.resample('Y').sum()
downsampled.index.rename('Yearly',inplace = True) 
downsampled.plot()
plt.title("Monthly Milk Production")
plt.xlabel("Months")
plt.show()
```

### Additive Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv('../Datasets/monthly-milk-production-pounds-p.csv')
series = df['Milk']
result = seasonal_decompose(series, model='additive', period=12)
result.plot()
plt.title('Additive Decompostion')
```

### Multiplicative Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv('../Datasets/monthly-milk-production-pounds-p.csv')
series = df['Milk']
result = seasonal_decompose(series, model='multiplicative', period=12)
result.plot()
plt.title('Multiplicative Decompostion')
```

## Shampoo Sales

```python
df = pd.read_csv('../Datasets/sales-of-shampoo-over-a-three-ye.csv', index_col =0)
df.index = pd.to_datetime(df.index).to_period("M")
```

```python
df.plot()
plt.title("Yearly Shampoo Sales")
plt.xlabel("Years")
plt.show()
```

### Quarterly

### Additive Decomposition
